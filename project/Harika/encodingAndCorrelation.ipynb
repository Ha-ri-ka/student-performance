```python
MLRCopy=data.copy()
MLRCopy.replace(['yes', 'no'], [1, 0], inplace=True) #BINARY ENCODING
MLRCopy.replace(['M', 'F'], [1, 0], inplace=True) #BINARY ENCODING
one_hot_encoded = pd.get_dummies(MLRCopy['school'], prefix='school') #one hot encoding
MLRCopy = pd.concat([one_hot_encoded, MLRCopy.drop('school', axis=1)], axis=1)
MLRCopy.replace(['R', 'U'], [1, 0], inplace=True) #BINARY ENCODING
MLRCopy.replace(['LE3', 'GT3'], [3, 4], inplace=True) #LABEL ENCODING
MLRCopy.replace(['A', 'T'], [1, 0], inplace=True) #BINARY ENCODING
MLRCopy.replace(['at_home', 'teacher', 'other','health','services'], [1,2,0,3,4], inplace=True) #custom encoding
MLRCopy.replace(['course','home','reputation'], [1,2,3], inplace=True)#custom encoding
MLRCopy.replace(['father', 'mother'], [1, 0], inplace=True) #custom encoding
```

```python
MLRCopy
```

FINDING CORRELATIONS OF EACH COLUMN WITH G3 TO DROP COLUMNS WITH WEAK CORRELATION

```python
correlation_matrix = MLRCopy.corr()['G3']
correlation_matrix
```

```python
cnt=0
for col,value in correlation_matrix.items():
    if abs(value)<0.1:
        cnt+=1
        print(f'{col} has a weak correlation with G3 ({value})')
print(cnt)
```
